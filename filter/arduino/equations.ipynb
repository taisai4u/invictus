```python
from sympy import symbols, Rational, Float

P, P0, h = symbols("P P0 h")
exp = Rational(1903, 10000)

# Rearrange manually: h = 44330*(1 - (P/P0)^exp)
# => (P/P0)^exp = 1 - h/44330
# => P = P0 * (1 - h/44330)^(1/exp)

P_solution = P0 * (1 - h / 44330) ** (1 / exp)
P_derivative = P_solution.diff(h)

P_solution, P_derivative
```

```python
from sympy import Matrix, sqrt
from sympy.core import symbol


ax, ay, az = symbols("a_x a_y a_z")
a = Matrix([ax, ay, az])
a_norm = sqrt(ax**2 + ay**2 + az**2)
G = a/a_norm
J = G.jacobian(Matrix([ax, ay, az]))
J.subs(a_norm, symbols("||a||"))
```

```python
from sympy import eye

v = a/a_norm
((eye(3) - v*v.T)/a_norm).subs(a_norm, symbols("||a||"))
```
